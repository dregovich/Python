# Описание данных

user_data:

client – идентификатор пользователя

premium – является ли клиент премиум

age – возраст

logs:

client – идентификатор пользователя

platform – платформа

```python
# Импортируем библиотеку pandas
import pandas as pd

# Загружаем данные
user_data = pd.read_csv('3_user_data.csv')
logs = pd.read_csv('3_logs.csv')

# Проверяем размер таблиц и типы переменных
print(f"Размер таблицы user_data: {user_data.shape}")
print(f"Типы переменных в user_data:\n{user_data.dtypes}")

print(f"Размер таблицы logs: {logs.shape}")
print(f"Типы переменных в logs:\n{logs.dtypes}")
```

# Задание 6

На основе этих выборок постройте графики, первый - отображающий популярность использования различных платформ (столбец "platform") у премиальных пользователей и второй - платформы сради непримиальных пользователей.

# Задание 7

Визуализируйте число успешных операций, сделанных на платформе computer, в зависимости от возраста, используя sns.countplot. По оси x – возраст, y – число успешных операций. 

Если размер графика получается небольшим, можно добавить plt.figure(figsize=(12, 8)) в начало соответствующей ячейки.

Подсказка: sns.countplot(data=Название_Датафрейма.loc[(Название_Датафрейма.platform == 'computer')
                                   & (Название_Датафрейма.success == True)], x='age')

```python
# Импортируем библиотеку pandas
import pandas as pd

# Загружаем данные
user_data = pd.read_csv('3_user_data.csv')
logs = pd.read_csv('3_logs.csv')

# Проверяем размер таблиц и типы переменных
print(f"Размер таблицы user_data: {user_data.shape}")
print(f"Типы переменных в user_data:\n{user_data.dtypes}")

print(f"Размер таблицы logs: {logs.shape}")
print(f"Типы переменных в logs:\n{logs.dtypes}")

# Задание 1
print(f"Количество уникальных значений в столбце 'platform': {logs['platform'].nunique()}")

# Задание 2
print("Платформа с наибольшим количеством успешных операций:")
print(logs.query('success == True').platform.value_counts().idxmax())

# Задание 3
print("Клиент с наибольшим количеством успешных операций:")
print(logs.query('success == True').client.value_counts().idxmax())

# Задание 4
combined_df = user_data.merge(logs, on='client')
print(f"Размер объединенного датафрейма: {combined_df.shape}")

# Задание 5
print("Платформа, наиболее популярная среди премиумных клиентов:")
premium_users = combined_df[combined_df['premium'] == True]
print(premium_users['platform'].value_counts().idxmax())

# Задание 6
premium_users = combined_df[combined_df['premium'] == True]
non_premium_users = combined_df[combined_df['premium'] == False]

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.title("Платформы среди премиумных пользователей")
premium_users['platform'].value_counts().plot(kind='bar')

plt.figure(figsize=(12, 6)) 
plt.title("Платформы среди непремиумных пользователей")
non_premium_users['platform'].value_counts().plot(kind='bar')

# Задание 7
plt.figure(figsize=(12, 8))
combined_df.loc[(combined_df.platform == 'computer') & (combined_df.success == True)]['age'].plot(kind='hist')
plt.title("Число успешных операций на платформе computer в зависимости от возраста")
plt.xlabel('Age')
plt.ylabel('Number of Successful Operations')
```
